```python
# Importing the libraries and Dataset
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris            # importing iris dataset
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
%matplotlib inline
```

```python
iris = load_iris()
```

```python
df= pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
df.head()
```

```python
df.shape
```

# Univariate

```python
df_setosa = df.loc[df["target"]==0]
```

```python
df_viginica = df.loc[df["target"]==1]
```

```python
df_versicular = df.loc[df["target"]==2]
```

```python
plt.plot(df_setosa["sepal length (cm)"], np.zeros_like(df_setosa["sepal length (cm)"]), 'o')
plt.plot(df_viginica["sepal length (cm)"], np.zeros_like(df_viginica["sepal length (cm)"]), 'o')
plt.plot(df_versicular["sepal length (cm)"], np.zeros_like(df_versicular["sepal length (cm)"]), 'o')
plt
plt.show()
```
