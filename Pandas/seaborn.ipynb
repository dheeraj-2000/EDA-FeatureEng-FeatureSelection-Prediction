```python
df = sns.load_dataset("tips")
```

```python
sns.jointplot(y = 'total_bill', x= 'tip', data = df, kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.distplot(df['tip'])
```

## Categorical feature

```python
sns.countplot(df['day'])
```

```python
sns.countplot(y=df['smoker'])
```

```python
sns.barplot(x = df['smoker'], y=df['total_bill'] )
```

```python
sns.boxplot(x = df['day'], y=df['total_bill'], palette='rainbow' )
```

```python
sns.boxplot(x = df['total_bill'], y=df['day'], hue= df['smoker'])
```

```python
sns.violinplot(x = df['total_bill'], y=df['day'])
```
