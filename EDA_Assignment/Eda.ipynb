## Overall Desktop and Mobile user for both Platform

```python
fig=plt.figure(figsize=(8,8))
sns.countplot(df['campaign_platform'], hue=df['device'])
plt.ylabel('Number of all types of Devices used', fontsize =15) 
plt.xlabel('Campaign Platform', fontsize =15)
# Add a legend
plt.title('Total number of devices used for each \nCompaign platform',fontweight="bold", fontsize =18)
plt.xticks(fontsize =13)
plt.yticks(fontsize =13)
plt.legend(loc='upper right', ncol=1, prop={"size":13})
plt.show()
```

## Compaign Platform vs total spends

```python
fig=plt.figure(figsize=(8,8))
color = ['green', 'blue']
df.groupby('campaign_platform').spends.sum().plot.bar(width = 0.3, color = color, ylim=0)
plt.ylabel('Total Spends in each Campaign', fontsize =15) 
plt.xlabel('Campaign Platform', fontsize =15)
# Add a legend
plt.title('Total Spends for each Compaign platform',fontweight="bold", fontsize =18)
plt.xticks(fontsize =13)
plt.yticks(fontsize =13)
# plt.legend(loc='upper left', ncol=1, prop={"size":13})
plt.show()
```
